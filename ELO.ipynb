```python
historic_elo=pd.DataFrame
historic_elo=pd.DataFrame(df['Athlete'].unique())
historic_elo['historic_elo'] = np.random.randint(1490,1510, size=len(historic_elo))
historic_elo
dict_elo=historic_elo.set_index(0).T
dict_elo
```

```python
for tävling in df['Event'].unique():
    
    df_iter=df[df['Event']==tävling]
    df_iter['Looser']=df_iter['Athlete'].shift(periods=-1)
    df_iter=df_iter.dropna()
    
    print(df_iter)
    for index, row in df_iter.iterrows():
    
        k_factor=32

        winner = row['Athlete']
        looser = row['Looser']

        winner_elo = dict_elo[winner]
        looser_elo = dict_elo[looser]

        prob_win_home = 1 / (1 + 10**((looser_elo - winner_elo) / 400))
        prob_win_away = 1 - prob_win_home

        new_winner_elo = winner_elo + k_factor*(1 - prob_win_home) * 1
        new_looser_elo = looser_elo + k_factor*(0 - prob_win_away) * 1

        tmp_looser = new_looser_elo

        dict_elo[winner] = new_winner_elo
        dict_elo[looser] = new_looser_elo
        print(dict_elo['JOHAUG Therese'])
        
dict_elo
   
```
